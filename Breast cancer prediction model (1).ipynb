TRAINING THE MODEL:

```python
prediction_vars=['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean',
                 'concave points_mean', 'radius_se', 'area_se',  'radius_worst',
                 'perimeter_worst', 'compactness_worst']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test=train_test_split(bc_data, test_size=0.15, random_state=1)
```

```python
train_x=train[prediction_vars]
train_y=train["diagnosis"]
test_x=test[prediction_vars]
test_y=test["diagnosis"]
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
model= RandomForestClassifier()
```

```python
model.fit(test_x, test_y)
```

```python
predictions= model.predict(test_x)
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predictions)
```

```python
print(test_y.value_counts())
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```python
precision=precision_score(test_y, predictions)
print("The precision is:", precision)

recall=recall_score(test_y, predictions)
print("The recall is:", recall)

accuracy=accuracy_score(test_y, predictions)
print("The accuracy is:", accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_y, predictions))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(test_y, predictions)
print("Mean Absolute Error (MAE):", mae)

rmse = mean_squared_error(test_y, predictions, squared=False)
print("Root Mean Squared Error (RMSE):", rmse)
```
